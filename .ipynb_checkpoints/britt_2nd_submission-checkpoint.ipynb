```python
import pandas as pd
```

```python
df = pd.read_csv('./datasets/train.csv')
```

```python
df.isnull().sum().head()
```

```python
df.shape
```

```python
df = pd.get_dummies(df, columns=['MS Zoning'], drop_first=True)
df.head()
```

```python
features = ['Lot Area', 'Overall Qual', 'Year Built', 'MS Zoning_C (all)', '1st Flr SF', '2nd Flr SF', 
            'Full Bath', 'Half Bath', 'Yr Sold']
X = df[features]
```

```python
X.head()
```

```python
y = df['SalePrice']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
```

```python
# instantiate
lr = LinearRegression()
```

```python
cross_val_score(lr, X_train, y_train)
```

```python
lr.fit(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

```python
import numpy as np
```

```python
np.mean(cross_val_score(lr, X_train, y_train))
```
