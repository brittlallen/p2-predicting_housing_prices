```python
url = 'https://api.ed.gov/data/crdc_oosuspensions_2013-14?&'
my_key = 'api_key=Euaf3lzPuB2l7KInxQxSkjtLjjyF7KGcJtfXdikJ&'
page_criteria = 'page='

pages = range(1,4777)
```

```python
master_df = pd.DataFrame()

for page_num in pages[:200]:
    
        print(url + my_key + page_criteria + str(page_num))
        response = requests.get(url + my_key + page_criteria + str(page_num))
        my_dict = response.json()
    
        df = pd.DataFrame(my_dict['resources'])
        master_df = master_df.append(df)
        #master_df = pd.concat([master_df, df]) -- another way to do the line just before
        print('digesting page... ' +str(page_num))
```
